```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_table("data.csv", delimiter = ',')
```

```python
years = data.iloc[:,0]
co2Emissions = data.iloc[:,1]
```

```python
plt.plot(years, co2Emissions)
```

```python
density = 100
mat = np.zeros([len(co2Emissions), density + 1])

for i in range(0, len(co2Emissions)):
    each = co2Emissions[i]
    sigma = 0.2 * each
    mu = each
    mat[i,1:] = sigma * np.random.randn(density) + mu

mat[:,0] = years
```

```python
plt.plot(years, mat[:,1:], 's', alpha = 0.1, color = '#ff89c0');
```

```python
subjectString = 'Date'
for i in range(0, mat.shape[1]):
    subjectString = subjectString + ',Subject_' + str(i)
```

```python
subjectString
```

```python
with open('userData.csv', 'wb') as f:
    f.write(subjectString + '\n')
    np.savetxt(f, mat, delimiter=",", fmt = '%.3e')
```
