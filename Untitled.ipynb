```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_table("data.csv", delimiter = ',')
```

```python
years = data.iloc[:,0]
co2Emissions = data.iloc[:,1]
```

```python
plt.plot(years, co2Emissions)
```

```python
density = 100
mat = np.zeros([len(co2Emissions), density])

for i in range(0, len(co2Emissions)):
    each = co2Emissions[i]
    sigma = 0.2 * each
    mu = each
    mat[i,:] = sigma * np.random.randn(density) + mu
```

```python
plt.plot(years, mat, 's', alpha = 0.1, color = '#ff89c0');
```

```python
np.savetxt("userData.csv", mat, delimiter=",", fmt = '%.3e')
```
